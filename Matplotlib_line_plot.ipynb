```python
# plot a graph by using a single co-ordinate
y = [2,4,1,5,8]
plt.plot(y)
plt.show()
```

```python
# change color and points on a graph
plt.plot(y,'ro')
plt.show
```

```python
plt.plot(y,'go--')
plt.show()
```

```python
# In a single graph pass a multiple keywords argument
plt.plot(y, color='green',marker='o',linestyle='dashed',linewidth=4,ms=12,mfc='r',mec='g')
plt.show()
```

```python
# axis labeling and plotting lines/grid in a graph
plt.plot(y,color='green',marker='o',linestyle='dashed',lw=4,ms=12,mfc='r',mec='g')
plt.xlabel("Salary")
plt.ylabel("Cars")
plt.title('Expense',loc='left')
plt.grid() # here also select a single axis
```
