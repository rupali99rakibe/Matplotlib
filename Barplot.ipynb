```python
import matplotlib.pyplot as plt
```

```python
x = ['A','B','C','D']
y = [8,5,11,7]
```

```python
# in a bar plot Two types vertical bar plot and horizontal bar plot by default it is a vartical
# To compare different categeorical or discrete variable such as age group 
plt.bar(x,y)
plt.show()
```

```python
# Horizontal bar plot
plt.barh(x,y)
plt.show()
```
