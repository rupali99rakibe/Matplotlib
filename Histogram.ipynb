```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.normal(150,20,200)
x
```

```python
# histogram use for a different ranges
plt.hist(x)
plt.show()
```

```python
# by default it can take bins but in a case we need to a change bins no. we can change
plt.hist(x,bins=20)
plt.show()
```

```python
# given a different bins size plotting a histogram
bins_size = [90,100,110,120,130,140,160,170,180,190,200]
plt.hist(x,bins=bins_size,orientation='horizontal',color='r')
plt.show()
```
