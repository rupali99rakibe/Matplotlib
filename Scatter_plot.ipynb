```python
import matplotlib.pyplot as plt
```

```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]
speed = [99,86,87,88,111,86,103,87,94,78,77,85,85]
speed1 =[85,99,111,86,97,94,87,78,11,55,99,77,15]
```

```python
len(age)
```

```python
len(speed)
```

```python
len(speed1)
```

```python
# scatter_plot analysis of a data(increment or decrement)
plt.scatter(age,speed)
plt.show()
```

```python
# one dependent co-ordinate and two independent co-ordinate
plt.scatter(age,speed,marker='*',color='r')
plt.scatter(age,speed1)
plt.show()
```

```python
# labelling on a graph
plt.scatter(age,speed,marker='*',color='r')
plt.scatter(age,speed1)
plt.xlabel('age')
plt.ylabel('speed of a car')
plt.show()
```

```python
# Title on a scatter plot
plt.scatter(age,speed,marker='*',color='r')
plt.scatter(age,speed1)
plt.xlabel('age')
plt.ylabel('speed of a car')
plt.title('analysis_speed_of_a_car')
plt.show()
```

```python
# grid plotting
plt.scatter(age,speed,marker='*',color='r')
plt.scatter(age,speed1)
plt.xlabel('age')
plt.ylabel('speed of a car')
plt.title('analysis_speed_of_a_car')
plt.grid()
plt.show()
```
