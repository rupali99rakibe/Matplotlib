```python
import matplotlib.pyplot as plt
```

```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]
speed = [99,86,87,88,111,86,103,87,94,78,77,85,85]
speed1 =[85,99,111,86,97,94,87,78,11,55,99,77,15]
```

```python
plt.scatter(age,speed,marker='*',c='r')
plt.scatter(age,speed)
plt.scatter(age,speed1)
plt.show()
```

```python
# legend
plt.scatter(age,speed,marker='*',c='r')
plt.scatter(age,speed)
plt.scatter(age,speed1)
plt.legend(['speed','speed1'])
plt.show()
```

```python
# multiple parameter in legend
plt.scatter(age,speed,marker='*',c='r')
plt.scatter(age,speed)
plt.scatter(age,speed1)
plt.legend(['speed','speed1'],loc='upper right',ncol=2,facecolor='yellow',fontsize=20,edgecolor='red')
plt.show()
```

```python
plt.scatter(age,speed,marker='*',c='r')
plt.scatter(age,speed)
plt.scatter(age,speed1)
plt.legend(['speed','speed1'],loc='upper right',ncol=2,facecolor='yellow',fontsize=20,edgecolor='red')
plt.xlabel('age',fontsize=20)
plt.ylabel('speed',fontsize=20)
plt.show()
```
