```python
import matplotlib.pyplot as plt
```

```python
# pie chart use for data analysis in percentage.
y = [40,10,20,30]
plt.pie(y)
plt.show()
```

```python
# change a color pie chart and give a name each part
y = [40,10,20,30]
x = ['A','B','C','D']
c = ['b','r','darkmagenta','k']
plt.pie(y,labels=x,colors=c)
plt.show()
```

```python
# Explodng (removing a particular section)
y = [40,10,20,30]
x = ['A','B','C','D']
c = ['b','r','darkmagenta','k']
plt.pie(y,labels=x,colors=c,explode=[0.0,0.1,0.0,0.0])
plt.show()
```

```python
# change an rotation angle
y = [40,10,20,30]
x = ['A','B','C','D']
c = ['b','r','darkmagenta','k']
plt.pie(y,labels=x,colors=c,startangle=90)
plt.show()
```

```python
# in shadow form
y = [40,10,20,30]
x = ['A','B','C','D']
c = ['b','r','darkmagenta','k']
plt.pie(y,labels=x,colors=c,startangle=90,shadow=True)
plt.show()
```
